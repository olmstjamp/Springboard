```python
import pandas as pd
import prince
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats as ss
import seaborn as sns
import warnings
warnings.filterwarnings(action='once')
%matplotlib inline
```

```python
data = pd.read_csv("CleanedData.csv", header=[0,1,2,3], index_col=0)
data.head()
```

```python
data.sum().groupby(by='Section').count()
```

```python
#This function just is a simplified way to get a slice from a multiindexed column
get_slice = lambda name, level: data.loc[:,data.columns.get_level_values(level)==name].droplevel(level=[1,2,3],axis=1)
```

```python
X = data.drop(['ID', 'Weight', 'RespInfo'], level=3, axis=1)
X  = X.astype(str)

sectionlist = X.columns.get_level_values(3).unique().sort_values()
groups = {section: list(get_slice(section,3).columns) for section in sectionlist}
X = X.droplevel(level=[1,2,3],axis=1)
X
```

```python
#Ran out of memory to try and extract anything from the mfa instance, so I just used a bunch of mcas to get components

mfa = prince.MFA(
        groups=groups,
        n_components=1,
        n_iter=3,
        copy=True,
        check_input=True,
        engine='auto',
        random_state=42)

mfa = mfa.fit(X)
```

```python
mfa.explained_inertia_ 
```

```python
#Using an automated elbow method to get the number of components for each section
print("Section n_components")
inertia_dict = {}
for section in sectionlist:
    Xsection = get_slice(section,3).astype(str)
    mca = prince.MCA(n_components=10)
    mca = mca.fit(Xsection)
    #print(section,mca.explained_inertia_)
    inertia_dict.update({section: mca.explained_inertia_})
    slopes = abs(np.diff(mca.explained_inertia_))
    #print(slopes, np.mean(slopes))
    #print(slopes>np.mean(slopes))
    cutoff = slopes[slopes>np.mean(slopes)][-1]
    best_n = np.where(slopes == cutoff)[0][0]+2
    print(section, best_n)
    
    mca = prince.MCA(n_components=best_n, random_state=111)
    mca = mca.fit(Xsection)
    
    newaddition = pd.DataFrame(data=np.array(mca.transform(Xsection)),\
                                     columns = [section+"_"+str(component_num) for component_num in range(1,best_n+1)])
    
    if section == sectionlist[0]:
        AllComponents = newaddition
    else:
        AllComponents = AllComponents.join(newaddition)

AllComponents.head()
```

```python
#The shape of the original data and the component data
print(AllComponents.shape, X.shape)
```

```python
corrmat = AllComponents.corr()
```

```python
#Correlation table of which components correspond to which question/answer combo. 
#eg. Clinton_Rubio_2016_1.0 means they answered a 1 to the original question titled Clinton_Rubio_2016.

def coltocomp_correlation(section,qa_labels=None,plot = False):
    section_components = AllComponents.loc[:,AllComponents.columns.str.startswith(section)]
    n = section_components.shape[1]
    corr = section_components.join(pd.get_dummies(get_slice(section,3).astype(str))).corr().iloc[n:,:n]
    
    if plot==False: return corr

    fsize_row = min(20,max(5,corr.shape[0]-6))
    fsize_col = corr.shape[0]/1.3
    plt.figure(figsize = (fsize_row,fsize_col))

    if qa_labels is not None:
        corr.index = qa_labels
    
    ax = sns.heatmap(corr, cmap="coolwarm", annot=True)
    plt.title(section+" Componets")
    plt.xlabel("")
    plt.ylabel("")
    plt.show()
    return
```

```python
test_section = "AmericanID"

AmericanID_Qs = ["Born in USA", "American Citizen", "Lived in USA Most of Life", "Speak English",\
                 "Christian", "Respect Institutions/Laws", "White", "Accept Diversity"]
AmericanID_As = ["Very", "Fairly", "Not Very", "Not at All", "DK/Blank"]
qa_interpretations = [q+" ("+a+")" for q in AmericanID_Qs for a in AmericanID_As]

print(test_section+" Explained_Inertia",inertia_dict[test_section])
coltocomp_correlation(test_section, plot=True)
```

```python
test = AllComponents.loc[:,AllComponents.columns.str.startswith(test_section)]
from sklearn.cluster import KMeans
Ks = range(2,20)
inertias = []

for K in Ks:
    kmeans_test = KMeans(n_clusters=K, random_state=111).fit(test)
    inertias.append(kmeans_test.inertia_)


plt.plot(Ks, inertias)
plt.title("Elbow Method Examination")
plt.xlabel("K")
plt.ylabel("SS")
plt.show()
```

```python
#Getting our final kmeans
bestK_test = 5
kmeans_test = KMeans(n_clusters=bestK_test, random_state=111).fit(test)
```

```python
def crosstab_label(colname, title, question_meanings):
    x = np.array(data.loc[:,data.columns.get_level_values(0)=='Label']).ravel()
    y = np.array(data.loc[:,data.columns.get_level_values(0)==colname]).ravel()
    table = pd.crosstab(x,y).divide(np.array(pd.crosstab(x,y).sum(axis=1)),axis=0)
    table.columns = question_meanings

    fsize_row = min(20,max(5,table.shape[0]-6))
    fsize_col = table.shape[0]/1.3
    plt.figure(figsize = (fsize_row,fsize_col))
    ax = sns.heatmap(table.T, cmap="hot", annot=True)
    plt.title(title)
    plt.xlabel("KMeans Labels")
    plt.show()
```

```python
q = ["Very Liberal", "Liberal", "Moderate", "Conservative", "Very Conservative", "Not Sure"]
crosstab_label("ideo5_2018","Liberal-Conservative Scale vs KMeans Labels",q)

#How each kgroup put themselves on a lib-con scale. Note each col adds up to 1
```

```python
def mean_diff_labels(colname, labela, labelb):
    groupa = pd.Series(data[data['Label'] == labela][colname].values.ravel())
    groupb = pd.Series(data[data['Label'] == labelb][colname].values.ravel())
    return np.mean(abs(groupa.value_counts(normalize=True).sort_index() - groupb.value_counts(normalize=True).sort_index()))

def mean_diff_labels_all(labela, labelb):
    
    diffs = pd.Series(index=X.columns)
    
    for i, colname in enumerate(X.columns):
        diffs[colname] = mean_diff_labels(colname, labela, labelb)

    return diffs
```

```python
group2_6 = mean_diff_labels_all(2, 6)
group1_5 = mean_diff_labels_all(1, 5)
```
