```python
def get_year(col):
    if col[-4:] == "2012":
        year = 2012
    elif col[-4:] == "2016":
       year = 2016
    elif col[-4:] == "2017":
        year = 2017
    elif col[-4:] == "2018":
        year = 2018
    elif col[-8:] == "baseline":
        year = 2011
    else:
        year = "Other"
    
    return year
```

```python
arrays = [np.array(data.columns),\
        np.array(pd.Series(data.columns).apply(get_year)),\
        np.array(coltypes['Type'])]

midx = pd.MultiIndex.from_arrays(arrays, names=('Col_name', 'Year', 'Type'))

data = pd.DataFrame(np.array(data), columns=midx)
```

```python
data.head()
```

```python
#Drop cols that aren't relevant
mask = (data.columns.get_level_values(2)!='Rand Indicator') & (data.columns.get_level_values(2)!='Probably drop')
data = data[data.columns[mask]]
```

```python
#Fill in the blanks of all Cat columns with 997 (representing 'Don't know' in other temp ratings)
data.loc[:, data.columns.get_level_values(2)=='Cat'] = data.loc[:, data.columns.get_level_values(2)=='Cat'].fillna(value=997)
print("Total %NA",data.isna().sum().sum()/(data.shape[0]*data.shape[1]))
```

```python
#Fill in the blanks of all 1-100 columns with the column average
avgs = data.loc[:, data.columns.get_level_values(2)=='1-100'].mean()
data.loc[:, data.columns.get_level_values(2)=='1-100'] = \
    data.loc[:, data.columns.get_level_values(2)=='1-100'].fillna(value=avgs)
print("Total %NA",data.isna().sum().sum()/(data.shape[0]*data.shape[1]))
```

```python
#Fill in the blanks of all 1-100_DN columns with 997
data.loc[:, data.columns.get_level_values(2)=='1-100_DN'] = \
    data.loc[:, data.columns.get_level_values(2)=='1-100_DN'].fillna(value=997)
print("Total %NA",data.isna().sum().sum()/(data.shape[0]*data.shape[1]))
```

```python
DN = data.loc[:, data.columns.get_level_values(2)=='Cat_O_DN'].max()
data.loc[:, data.columns.get_level_values(2)=='Cat_O_DN'] = \
    data.loc[:, data.columns.get_level_values(2)=='Cat_O_DN'].fillna(value=DN)
print("Total %NA",data.isna().sum().sum()/(data.shape[0]*data.shape[1]))
```

```python
#free trade needs 2 and 3 switched, but is then an Cat_O
#tax, regula, house_amount needs 2 and 3 switched, but is then a Cat_O_DN
#localnewspaper_baseline needs all 9 to switch to 0 (# of days/week)
```

```python
#Swtiching the 2 and 3 responses to make the questions Cat_O_DN for all the similiarly structured questions
switch_23_list = list(data.columns.get_level_values(0)[data.columns.get_level_values(2)=='Change order'])[:-1]
print(switch_23_list[:9])
data.loc[:,data.columns.get_level_values(0).isin(switch_23_list)] =\
    data.loc[:,data.columns.get_level_values(0).isin(switch_23_list)].replace(2,444)

data.loc[:,data.columns.get_level_values(0).isin(switch_23_list)] =\
    data.loc[:,data.columns.get_level_values(0).isin(switch_23_list)].replace(3,2)

data.loc[:,data.columns.get_level_values(0).isin(switch_23_list)] =\
    data.loc[:,data.columns.get_level_values(0).isin(switch_23_list)].replace(444,3)

#Changing all 9s to 0s in the localnewspaper_baseline question to represent 0 days of a week reading the paper
data.loc[:,data.columns.get_level_values(0)=='localnewspaper_baseline'] =\
    data.loc[:,data.columns.get_level_values(0)=='localnewspaper_baseline'].replace(9,0)
```
